```python
import pandas as pd
X_train = pd.read_csv("./johannes_working_ml/Drug1803_analysis/X_data.csv")
X_test = pd.read_csv("./johannes_working_ml/Drug1803_analysis/X_test.csv")
y_train = pd.read_csv("./johannes_working_ml/Drug1803_analysis/labels.csv", header= None)
y_test = pd.read_csv("./johannes_working_ml/Drug1803_analysis/test_labels.csv", header=None)
```

```python
X_train.shape
```

```python
from xgboost import XGBRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import r2_score, mean_squared_error
from sklearn.linear_model import ElasticNet
import numpy as np

def recurse_pred(train_features = X_train, test_features = X_test, mask = np.ones(X_train.shape[-1], dtype=np.bool_), threshold = 1e-5, depth = 0):
    if (mask.sum() == 0) or (depth != 0 and train_features.shape[-1] == mask.sum()): return;
    
    train_features = train_features.loc[:, mask]
    test_features = test_features.loc[:, mask]
    print(f"Evaluate {train_features.shape[-1]} features.")

    model = RandomForestRegressor(n_jobs = -1) #XGBRegressor(**{'device': 'cuda:0', 'seed': 0, 'tree_method': 'hist'})
    model.fit(train_features.values, y_train.values)

    r2_eval = r2_score(y_test, model.predict(test_features))
    print(f"R2-score: {r2_eval:.4f}")
    feature_importance = model.feature_importances_ if hasattr(model, 'feature_importances_') else model.coef_[0]
    mask = feature_importance > threshold
    depth += 1
    recurse_pred(train_features, test_features, mask, threshold, depth)
```

```python
recurse_pred()
```

```python
X_train = X_train.loc[:, ~mask]
X_test = X_test.loc[:, ~mask]
X_train.shape, X_test.shape
```

```python
from xgboost import XGBRegressor

xbr = XGBRegressor(**{'alpha': 0.001,
 'colsample_bytree': 0.8,
 'device': 'cuda:0',
 'eta': 0.2,
 'gamma': 0.1,
 'max_depth': 6,
 'seed': 0,
 'tree_method': 'hist'})
xbr.fit(X_train.values, y_train.values)
```

```python
from sklearn.metrics import r2_score, mean_squared_error
```

```python
from sklearn.model_selection import GridSearchCV

parameters = {
                "tree_method": ["hist"],
                "device":["cuda:0"],
                "eta": (0.2, 0.3, 0.4),
                "gamma": (0.0, 0.01, 0.1),
                "max_depth": (4, 6, 8),
                "colsample_bytree": (0.8, 0.9, 1),
                "alpha": (0, 1e-3, 1e-1), ##l1 penalty
                "seed": [0], ## exclude randomizations
             }
clf = GridSearchCV(xbr, parameters, scoring = "r2", n_jobs=5, refit = True, cv = 5, verbose = 2.1, return_train_score = False)
clf.fit(X_train.values, y_train.values)
```

```python
from sklearn.metrics import r2_score, mean_squared_error
r2_score(y_test, clf.predict(X_test))
```
