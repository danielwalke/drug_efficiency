```python
import torch
from torch.nn import Conv1d
class LinearModel(torch.nn.Module):
    def __init__(self, hidden_dims, dropouts):
        super(LinearModel, self).__init__()
        self.lin_layers = torch.nn.ModuleList()
        for i in range(len(hidden_dims) - 1):
            self.lin_layers.append(torch.nn.Linear(hidden_dims[i], hidden_dims[i+1]))
        self.dropouts = [torch.nn.Dropout(dropout) for dropout in dropouts]
    
    def forward(self, x):
        for i, lin_layer in enumerate(self.lin_layers[:-1]):
            x = self.dropouts[i](x)
            x = lin_layer(x)
            x = torch.relu(x)
        return self.lin_layers[-1](x)
```

```python
from torch_geometric.data import HeteroData

def sparse_table_to_graph(feature_table_df, label_table_df, nodes_offset, is_train, embed_cell_models = None, embed_genes=None):
    gene_idx = torch.arange(feature_table_df.shape[-1])
    gene_enc = torch.zeros(feature_table_df.shape[-1], feature_table_df.shape[-1])
    gene_enc[gene_idx, gene_idx] = 1

    # if embed_genes is not None:
    #     gene_enc = gene_idx
        

    test_offset = 0 if is_train else nodes_offset
    idx = torch.arange(feature_table_df.shape[0])
    cell_model_enc = torch.zeros(feature_table_df.shape[0], feature_table_df.shape[0] + nodes_offset)
    cell_model_enc[idx, idx+test_offset] = 1

    # if embed_cell_models:
    #     cell_model_enc = idx+test_offset

    mask = feature_table_df.values.flatten() != 0
    edge_attributes = feature_table_df.values.flatten()[mask]

    #torch.from_numpy(mask).unsqueeze(0).repeat((2, 1))
    edge_index = torch.cat([idx.unsqueeze(1).repeat((1,feature_table_df.shape[-1])).flatten().unsqueeze(0), gene_idx.repeat(feature_table_df.shape[0]).unsqueeze(0)], axis = 0)[:, mask]
    edge_index_rev = torch.clone(edge_index)
    edge_index_rev[1] = torch.clone(edge_index)[0]
    edge_index_rev[0] = torch.clone(edge_index)[1]

    data = HeteroData()
    data['model'].x = cell_model_enc
    data['gene'].x = gene_enc
    data['model', 'expression', 'gene'].edge_index = edge_index
    data['model', 'expression', 'gene'].edge_attr = edge_attributes
    data['gene', 'expression', 'model'].edge_index = edge_index_rev
    data['gene', 'expression', 'model'].edge_attr = edge_attributes
    
    y = torch.from_numpy(label_table_df.values).type(torch.float)
    data['model'].y = y
    return data
```

```python
device = torch.device("cuda:2")
```

```python
from torch_geometric.data import HeteroData

train_data = sparse_table_to_graph(X_train_exp, y_train, X_test_exp.shape[0], True).to(device)
test_data = sparse_table_to_graph(X_test_exp, y_test, X_train_exp.shape[0], False).to(device)
```

```python
from torch_geometric.nn import GATConv, Linear
class GAT(torch.nn.Module):
    def __init__(self,hidden_dim = 16, heads = 1):
        super(GAT, self).__init__()
        self.lin = Linear(-1, hidden_dim)
        self.gat = GATConv((-1,-1), hidden_dim, heads = heads,concat = False, add_self_loops=False)
        self.gat_out = GATConv(hidden_dim, 1, heads = heads,concat = False, add_self_loops=False)
    def forward(self, x, edge_index, edge_attr):
        x = self.lin(x)
        # x = self.gat(x, edge_index, edge_attr)
        x = self.gat_out(x, edge_index, edge_attr)
        return x
```

```python
from torch_geometric.nn import to_hetero

lr = 1e-2
wd = 0
epochs = 5_000
model = GAT(4)
model = to_hetero(model, train_data.metadata(), aggr='mean').to(device)
optim = torch.optim.Adam(params=model.parameters(), lr= lr, weight_decay=wd)
loss_fn = torch.nn.MSELoss()
```

```python
from sklearn.metrics import r2_score, root_mean_squared_error
def evaluate(data):
    with torch.inference_mode():
        model.eval()
        out = model(data.x_dict, data.edge_index_dict, data.edge_attr_dict)
        r2 = r2_score(data["model"].y.cpu().numpy(), out["model"].cpu().numpy())
        rmse = root_mean_squared_error(data["model"].y.cpu(), out["model"].cpu().numpy())
    return r2, rmse
```

```python
for epoch in range(epochs):
    model.train()
    optim.zero_grad()
    out = model(train_data.x_dict, train_data.edge_index_dict, train_data.edge_attr_dict)
    loss = loss_fn(out["model"], train_data["model"].y)
    train_scores, test_scores = evaluate(train_data), evaluate(test_data)
    r2_train, rmse_train = train_scores
    r2_eval, rmse_eval = test_scores 
    print(f"{epoch}: {r2_train=:.4f}, {rmse_train=:.4f}, {r2_eval=:.4f}, {rmse_eval=:.4f}")
    loss.backward(retain_graph=True)
    optim.step()
```

```python
import torch
from torch.nn import Parameter
from torch_geometric.nn import MessagePassing
from torch_geometric.utils import add_self_loops, degree
from torch_geometric.nn import Linear
from torch_geometric.utils import degree

# This function is copied from https://github.com/gordicaleksa/pytorch-GAT/blob/main/The%20Annotated%20GAT%20(PPI).ipynb
## Thanks to Aleksa Gordic for the implementation!
def explicit_broadcast(this, other):
    # Append singleton dimensions until this.dim() == other.dim()
    for _ in range(this.dim(), other.dim()):
        this = this.unsqueeze(-1)

    # Explicitly expand so that shapes are the same
    return this.expand_as(other)

def describe(tensor):
    return f"{tensor.mean()} +- {tensor.std()} [{tensor.min()}, {tensor.max()}]"
    
class HeteroCustomGNN(torch.nn.Module):
    
    def __init__(self, out_channels, x_dict, emb_dim):
        ## message aggregator
        super().__init__()

        self.lin_emb = torch.nn.ModuleDict()
        for key in x_dict:
            self.lin_emb[key] = torch.nn.Linear(x_dict[key].shape[-1], emb_dim)

        self.out_dict = torch.nn.ModuleDict()
        for key in x_dict:
            self.out_dict[key] = torch.nn.Linear(emb_dim+1, out_channels)

    def forward(self, x_dict, edge_index_dict, edge_attr_dict):
        ##Message generation
        for key in x_dict:
            x_dict[key] = self.lin_emb[key](x_dict[key])
        message_dict = dict()
        for key in edge_index_dict:
            source, edge, target = key
            edge_index = edge_index_dict[key]
            
            source_lift = torch.index_select(x_dict[source], 0, edge_index[0])
            target_lift = torch.index_select(x_dict[target], 0, edge_index[1])
                   
            messages = torch.from_numpy(edge_attr_dict[key]).type(torch.float).to(device)
            messages = messages* source_lift.sum(-1) #torch.cat([source_lift, edge_attr_dict[key], target_lift], dim = 1)
            aggregated_messages = torch.zeros((x_dict[target].shape[0]), device = device, dtype=x_dict[target].dtype)
            
            broadcasted_target_index = explicit_broadcast(edge_index[1], messages)
            
            aggregated_messages.scatter_reduce_(0, broadcasted_target_index, messages, "sum")
            message_dict[key] = aggregated_messages.unsqueeze(1)
            
        # ## Message aggregation
        message_aggr_dict = dict()
        for key in message_dict:
            source, edge, target = key
            messages_for_target = message_dict[key]
            previous_message_aggr_target = message_aggr_dict.get(target, None)
            if previous_message_aggr_target is None:
                message_aggr_dict[target] = messages_for_target
            else:
                message_aggr_dict[target] = torch.cat((previous_message_aggr_target, messages_for_target), dim = 1)#previous_message_aggr_target + messages_for_target

        # x_dict["model"] = torch.cat((message_aggr_dict["model"], x_dict["model"]), dim = 1)
        x_dict["model"] = message_aggr_dict["model"] #self.out_dict["model"](x_dict["model"])
        return x_dict
```
